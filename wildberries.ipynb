```python
from bs4 import BeautifulSoup
```

```python
def search(types, response):
    element = response.select_one(types)
    return element.text.strip()
```

```python
path = '/Users/evgeniavu/Desktop/StudyHSE/Второй курс/ФЯ/parsing-with-python/tests/wildberries/tests_source/test0.html'
with open(path) as fp:
    response = BeautifulSoup(fp, 'html.parser')
    cards = response.select(r'.product-card')
    for card in cards:
        img_url = search(".product-card__img-wrap", card)
        brand = search(".product-card__brand", card)
        prices = search(".price-commission", card)
        print(img_url, brand, prices)
```
