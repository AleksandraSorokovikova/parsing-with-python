```python
from bs4 import BeautifulSoup
```

```python
def search_list(types, response):
    element = response.select(types)
    return [x.text.strip().lower() for x in element] 
```

```python
def search(types, response):
    element = response.select_one(types)
    return element.text.strip()
```

```python
def parse(file) :
    response = BeautifulSoup(file, 'html.parser')
    times = search(".tm-article-snippet__datetime-published", response)
    tags = search_list(".tm-tags-list__link", response)
    habs = search_list(".tm-hubs-list__link", response)
    saved = search(".bookmarks-button__counter", response)
    return {'times' : times, 'tags': tags, 'habs' : habs, 'saved' : saved}
```

```python
path = 'tests/habr/tests_source/test{0}.html'
for cnt in range(47):
    with open(path.format(cnt)) as fp:
        print(parse(fp))
```
